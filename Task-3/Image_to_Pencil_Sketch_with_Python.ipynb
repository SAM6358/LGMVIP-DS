```python
import matplotlib.pyplot as plt
import cv2
```

```python
original_img = cv2.imread('/content/Iron-man.webp')
```

```python
grayscale_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
```

```python
invert_img = cv2.bitwise_not(grayscale_img)
```

```python
blur_img = cv2.GaussianBlur(invert_img, (71,71) ,0)
```

```python
invert_blur_img = cv2.bitwise_not(blur_img)
```

```python
sketch_img = cv2.divide(grayscale_img,invert_blur_img,scale=256.0)
```

```python
plt.figure(figsize=(20,20))

plt.subplot(1,6,1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))

plt.subplot(1,6,2)
plt.title("GrayScale Image")
plt.axis("off")
plt.imshow(grayscale_img,cmap="gray")

plt.subplot(1,6,3)
plt.title("Inverted Image")
plt.axis("off")
plt.imshow(invert_img,cmap="gray")

plt.subplot(1,6,4)
plt.title("Blurred Image")
plt.axis("off")
plt.imshow(blur_img,cmap="gray")

plt.subplot(1,6,5)
plt.title("Inverted Blurred Image")
plt.axis("off")
plt.imshow(invert_blur_img,cmap="gray")

plt.subplot(1,6,6)
plt.title("Sketched Image")
plt.axis("off")
plt.imshow(sketch_img,cmap="gray")
```
