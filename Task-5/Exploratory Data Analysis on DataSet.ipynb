```python
#Correcting spellings of the columns of the Dataset

df_train.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'Latitude',
    'longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'Kill',
     'nwound':'Wound'},inplace=True)
```

```python
df_train.head()
```

```python
df_train.columns
```

```python
#Removig Unnecessary columns
df_train = df_train[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude',"Attacktype",'Kill',
               'Wound','Summary','Group','Targettype','Weapon','motive','target1']]
```

```python
print('NA percenage',df_train.isnull().sum()/df_train.isnull().count()*100)
```

```python
df_train['Kill'] = df_train['Kill'].fillna(0)
df_train['Wound'] = df_train['Wound'].fillna(0)
df_train['Casualties'] = df_train['Kill'] + df_train['Wound'] 
```

```python
year = df_train['Year'].unique()
years_count = df_train['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,y = years_count,palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Global Terrorists Attacks in Years',fontsize=30)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(df_train['Country'].value_counts()[:20].index,df_train['Country'].value_counts()[:20].values,palette='PuBuGn')
plt.title('Top Countries Affected By Terrorists Attacks',fontsize=30)
plt.xlabel('Countries', fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.xticks(rotation = 50)
plt.show()
```

```python
df_train.Group.value_counts()[:50]
```

```python
pd.crosstab(df_train.Year, df_train.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region',fontsize=30)
plt.ylabel('Number of Attacks By Terrorists',fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.show()
```

```python
df_train['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(20).plot(kind='bar',figsize=(20,10),color='red')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=20)
plt.ylabel("Number of attacks",fontsize=20)
plt.title("Top 20 most effected city",fontsize=30)
plt.show()
```

```python
df = df_train[['Year','Kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4,color='cyan')
plt.xticks(rotation = 50)
plt.title("People Died Due To Terrorist Attacks",fontsize=30)
plt.ylabel("Number of people killed",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
df_train['Attacktype'].value_counts().plot(kind='bar',figsize=(20,10),color='Purple')
plt.xticks(rotation = 50)
plt.xlabel("Terrorists AttackType",fontsize=20)
plt.ylabel("Number of Terrorist attacks",fontsize=20)
plt.title("Name of Terrorist Attacks",fontsize=30)
plt.show()
```

```python
df_train[['Attacktype','Kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['Green'])
plt.xticks(rotation=50)
plt.title("Number of killed People",fontsize=30)
plt.ylabel('Number of people',fontsize=20)
plt.xlabel('Attack type',fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df_train["Targettype"],order=df_train['Targettype'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("mako"));
plt.xticks(rotation=90)
plt.xlabel("AttackTypes",fontsize=20)
plt.title("Terrorists Attacks per year",fontsize=30)
plt.show()
```

```python
df_train[['Group','Kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Kill',ascending=False).head().plot(kind='bar',color='red',figsize=(20,10))
plt.title("Top 5 Terrorist group attack",fontsize=20)
plt.xlabel("Terrorist Group names",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
